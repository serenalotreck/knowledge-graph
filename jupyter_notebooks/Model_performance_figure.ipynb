## Figure including PURE


```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
```

```python
perf = pd.read_csv('../data/straying_off_topic_data/model_performance/all_gold_standard_abstracts_10Jan2022_performance_with_PURE.csv')
perf
```

```python
shorthand = ['ace05', 'scierc-light', 'genia-light', 'genia', 'scierc', 'phrasematched_plain', 'phrasematched_refined', 'ace05_pure', 'scierc_pure']
perf['shorthand'] = shorthand

colors = ['royalblue', 'royalblue', 'royalblue', 'royalblue', 'royalblue', 'coral', 'coral', 'orchid', 'orchid']
perf['colors'] = colors
```

```python
perf = perf.sort_values(by='F1', ascending=False)
perf
```

```python
# Custom legend extras
custom_lines = [mpatches.Patch(color='royalblue', label='DyGIE++'),
                mpatches.Patch(color='orchid', label='PURE'),
                mpatches.Patch(color='coral', label='Non-neural benchmark method')]

plt.rcParams['legend.handlelength'] = 1
plt.rcParams['legend.handleheight'] = 1.125
plt.rcParams['legend.numpoints'] = 1
```

```python
plt.bar(perf.shorthand, perf.F1, color=perf.colors)
plt.xticks(rotation=90)
plt.xlabel("Model", fontsize=14)
plt.ylabel("F1 Score", fontsize=14)
plt.title("Named Entity Recognition Model Performance", fontsize=20, pad=20)
plt.legend(handles=custom_lines)
```
