## Figure for PRISMA poster

Performance values:

```python
perf = {'dygiepp_scierc':0.41221374045801523,
       'dygiepp_scierc_light':0.4038461538461538,
       'dygiepp_genia':0.21527777777777782,
       'dygiepp_genia_light':0.17843866171003717,
       'dygiepp_ace05':0.13924050632911392,
       'phrasematch_plain':0.03773584905660377,
       'phrasematch_refined':0.033802816901408454}
```

Plot:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
```

```python
# Custom legend extras
custom_lines = [mpatches.Patch(color='purple', label='Neural method'),
                   mpatches.Patch(color='teal', label='Non-neural benchmark method')]

plt.rcParams['legend.handlelength'] = 1
plt.rcParams['legend.handleheight'] = 1.125
plt.rcParams['legend.numpoints'] = 1
```

```python
plt.bar(perf.keys(), perf.values(), color=['purple', 'purple', 'purple', 'purple', 'purple', 'teal', 'teal'])
plt.xticks(rotation=90)
plt.xlabel("Model", fontsize=14)
plt.ylabel("F1 Score", fontsize=14)
plt.title("Named Entity Recognition Model Performance", fontsize=20, pad=20)
plt.legend(handles=custom_lines)
```
